```python
import requests
from IPython import display
from base64 import b64decode
```

```python
# baseUrl = "https://p882bl70hqiua5-64410b71-3000.proxy.runpod.io/"
baseUrl = "https://localhost:7860/"
```

```python
models = requests.get(f'{baseUrl}sdapi/v1/sd-models').json()
```

```python
models
```

```python
styles = requests.get(f'{baseUrl}sdapi/v1/prompt-styles').json()
```

```python
styles
```

```python
# Call txt2img
# Notice I'm using a style. The prompt is just injected into the style string
# You should specify model everytime with "sd_model_checkpoint". Concepts will use different models that share the same name. The first model is the model for all collections.

prompt = "a basketball"
style = styles[1]['name']

data = {
  "enable_hr": False,
  "prompt": prompt,
  "negative_prompt": "",
  "styles": [
      style
  ],
  "seed": -1,
  "batch_size": 8,
  "n_iter": 1,
  "steps": 50,
  "cfg_scale": 10,
  "restore_faces": False,
  "override_settings": {"sd_model_checkpoint": models[0]['title']},
  "sampler_index": "Euler a"
}
r_txt2img = requests.post(f'{baseUrl}sdapi/v1/txt2img', json=data).json()
```

```python
## Check Progress

requests.get(f'{baseUrl}sdapi/v1/progress?skip_current_image=false').json()
```

```python
## Show images

for i in r_txt2img['images']:
    display.display(display.Image(b64decode(i)))
```

```python
# img2img
# The vary boolean is used for the vary vial. This will return images that are similar to the inputted image
# For some reason you have to add ':,' before the image

vary = True

data = {
  "init_images": [
    ';,' + r_txt2img['images'][0]
  ],
  "denoising_strength": 0.75,
  "mask_blur": 4,
  "inpainting_fill": 0,
  "inpaint_full_res": True,
  "prompt": prompt,
  "styles": [
      style
  ],
  "seed": -1,
  "subseed": -1 if not vary else 1,
  "subseed_strength": -1 if not vary else 1,
  "batch_size": 4,
  "n_iter": 1,
  "steps": 50,
  "cfg_scale": 10,
  "restore_faces": False,
  "negative_prompt": "",
  "override_settings": {},
  "sampler_index": "Euler a",
}
r_img2img = requests.post(f'{baseUrl}sdapi/v1/img2img', json=data).json()
```

```python
display.display(display.Image(b64decode(r_txt2img['images'][0])))
```

```python
for i in r_img2img['images']:
    display.display(display.Image(b64decode(i)))
```
